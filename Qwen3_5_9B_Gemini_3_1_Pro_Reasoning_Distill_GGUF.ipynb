```python
from llama_cpp import Llama

# Download and Load the model
# We set n_gpu_layers=-1 to ensure the T4 handles the entire workload
llm = Llama.from_pretrained(
    repo_id="Jackrong/Qwen3.5-9B-Gemini-3.1-Pro-Reasoning-Distill-GGUF",
    filename="Qwen3.5-9B.Q5_K_M.gguf",
    n_gpu_layers=-1,
    n_ctx=4096,      # Context window
    flash_attn=True  # Extra speed boost for the T4
)
```

```python
# Change the prompt below to whatever you want to ask!
user_query = "choose a nymber between 1 and 1 million"

# Start the completion
response = llm.create_chat_completion(
    messages=[
        {"role": "user", "content": user_query}
    ],
    stop=["<|im_end|>"],
    max_tokens=1024,
    stream=True
)

print("RoyalAI is thinking...\n" + "="*30 + "\n")

for chunk in response:
    if 'content' in chunk['choices'][0]['delta']:
        print(chunk['choices'][0]['delta']['content'], end='', flush=True)
```
